```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df5=pd.read_csv("C:/Users/athar/OneDrive/Desktop/Practical/DSPrac/HousingData.csv")
df5
```

```python
null_cols=df5.isnull().sum()
missing_cols = [col for col, count in null_cols.items() if count > 0]
for col in missing_cols:
    df5[col] = df5[col].fillna(df5[col].mean())

df5.isnull().sum()
```

```python
df5.describe()
```

```python
reg = LinearRegression()
x = df5.drop(columns=['MEDV'])
y = df5['MEDV']
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=0.20, random_state=123, shuffle=True)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
reg.fit(x_train, y_train)
```

```python
y_predict=reg.predict(x_test)
y_predict
```

```python
y_test
```

```python
reg.score(x_test, y_test)
```

```python
y_predict=reg.predict(x_test)
rms = (np.sqrt(mean_squared_error(y_test,y_predict)))
r2=round(reg.score(x_test, y_test),2)

print("The model of performance for training set")
print("\n")
print(f"Root Mean Squared Error is {rms}")
print(f"R^2 is {r2}")
```
