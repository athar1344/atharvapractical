```python
numeric_cols = df10.select_dtypes(include=np.number).columns
for col in numeric_cols:
    q1 = df10[col].quantile(0.25)
    q3 = df10[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR
    df10 = df10[(df10[col]>=lower_bound)&(df10[col]<=upper_bound)]

df10
```

```python
sns.boxplot(data=df10)
plt.title("Boxplot of Features in the Iris Dataset", fontsize=14)
plt.show()
```
